```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from lxml import etree
import yfinance as yf
```

```python
ticker = "AURE3"
base_url = f"https://www.dadosdemercado.com.br/acoes/{ticker}"
try:
    response = requests.get(base_url)
    response.raise_for_status()
except requests.exceptions.RequestException as e:
    print(f"Erro ao acessar página principal de ELET3: {e}")

soup = BeautifulSoup(response.text, 'html.parser')
tables = soup.find_all(attrs={'id': ('balances', 'cashflows', 'incomesYear')})
```

```python
i = 0
for idx, table in enumerate(tables):
    rows = table.find_all('tr')
    table_data = []

    for row in rows:
        headers_in_row = row.find_all('th')
        if headers_in_row:
            cols = [ele.text.strip() for ele in headers_in_row]
        else:
            cols = [ele.text.strip() for ele in row.find_all('td')]
        table_data.append(cols)
    var = 'df' + str(i)
    vars()[var] = pd.DataFrame(table_data)
    i += 1
```

```python
for i in range(0,3):
    var = 'df' + str(i)
    vars()[var].columns = vars()[var].iloc[0]
    vars()[var] = vars()[var][1:].reset_index(drop=True)
```

```python
def convert_value(value):
    multp = {'t': 1e12, 'b': 1e9, 'm': 1e6, 'mil': 1e3}
    val_str = str(value).strip().lower()
    parts = val_str.split(' ')

    if len(parts) == 2:
        x, y = parts
        x = x.replace(',', '.')
        x = float(x) * multp.get(y, 1)
        return int(x)
    
    return int(value)

try:
    for i in range(0,3):
        var = 'df' + str(i)
        vars()[var][vars()[var].columns[1:]] = vars()[var][vars()[var].columns[1:]].map(convert_value)
except Exception as e:
    print(f'Valores inconsistenmtes para o valuation de {ticker}: {str(e)}')
```

```python
def get_info(ticker_base):
    variations = [
        ticker_base,
        f"{ticker_base}.SA"
    ]
    
    for variation in variations:
        try:
            dat = yf.Ticker(variation)
            info = dat.info
            beta = info.get('beta')
            if beta is not None:
                print(f"Info from {ticker_base} successfully retrieved")
                return info
        except Exception:
            continue

    return None
```

```python
info = get_info(ticker)
beta = info.get('beta')
####################################
url = 'https://api.bcb.gov.br/dados/serie/bcdata.sgs.1178/dados/ultimos/1?formato=json'
resp = requests.get(url)
selic = float(resp.json()[0].get('valor')) + 0.1
####################################
Mkt_Price = 0.05
####################################
ke = (beta * Mkt_Price) + (selic/100)
ke
```

```python
Div_Liq = (df0_1[df0_1.columns[1]][14] + df0_1[df0_1.columns[1]][16]) - df0_1[df0_1.columns[1]][2]
P_Liq = df0_1[df0_1.columns[1]][17]
W_Div = Div_Liq / (Div_Liq + P_Liq)
W_P_Liq = P_Liq / (Div_Liq + P_Liq)
WACC = (selic/100) * W_Div * (1 - Tx_Imp) + W_P_Liq * ke
WACC
```

```python
def sect_tx(info):
        setores = {
            'Basic Materials': 0.031,
            'Communication Services': 0.081,
            'Consumer Cyclical': 0.047,
            'Consumer Defensive': 0.035,
            'Energy': 0.034,
            'Financial Services': 0.11,
            'Healthcare': 0.053,
            'Industrials': 0.04,
            'Real Estate': 0.103,
            'Technology': 0.109,
            'Utilities': 0.08
        }
        sector = info.get('sector')
        tx = setores.get(sector, 0.01)
        return tx
YG = sect_tx(info)
LT_G = YG/2
YG
```

```python
P_FCFF = [FCFF*(1+YG)]
for i in range(0,4):
    P_FCFF.append(P_FCFF[i]*(1+YG))
P_FCFF
```

```python
present_value = []
for i in range(0,5):
    present_value.append(P_FCFF[i]/(1+WACC)**(i+1))

present_value[4]
```

```python
Val_Terminal = P_FCFF[4]*(1+LT_G)/(WACC-LT_G)
ValP_Terminal = Val_Terminal/(1+WACC)**5
Val_Presente_Total = sum(present_value)+ValP_Terminal
Val_Presente_Total
```
