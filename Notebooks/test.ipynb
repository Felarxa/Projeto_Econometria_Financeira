```python
def fetch_html(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        return response.text
    except requests.exceptions.RequestException as e:
        print(f"Erro ao acessar página de Ações listadas na B3: {e}")
        return None

def extract_tables(html):
    soup = BeautifulSoup(html, 'html.parser')
    tables = soup.find_all(attrs={'id': 'stocks'})
    return tables

def parse_table(table):
    rows = table.find_all('tr')
    table_data = []

    for row in rows:
        headers = row.find_all('th')
        if headers:
            cols = [cell.text.strip() for cell in headers]
        else:
            cols = [cell.text.strip() for cell in row.find_all('td')]
        table_data.append(cols)

    return pd.DataFrame(table_data)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def parse_table(table):
        rows = table.find_all('tr')
        table_data = []

        for row in rows:
            headers = row.find_all('th')
            if headers:
                cols = [cell.text.strip() for cell in headers]
            else:
                cols = [cell.text.strip() for cell in row.find_all('td')]
            table_data.append(cols)

        return pd.DataFrame(table_data)

def extract_tables(html, ids):
        soup = BeautifulSoup(html, 'html.parser')
        tables = soup.find_all(attrs={'id': (ids)})
        return tables

def fetch_html(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        return response.text
    except requests.exceptions.RequestException as e:
        print(f"Erro ao acessar página: {e}")
        return None
    
def get_tickers():
        html = fetch_html('https://www.dadosdemercado.com.br/acoes')
        tables = extract_tables(html, 'stocks')
        df = [parse_table(table) for table in tables]
        df = df[0]
        df.columns = df.iloc[0]
        df = df[1:].reset_index(drop=True)
        df_filtrado = df[df['Negócios'].str.len() > 8]
        df_filtrado.drop(df_filtrado.columns[1:], axis=1, inplace=True)
        tickers_list = df_filtrado['Ticker'].tolist()
        return tickers_list

tickers = get_tickers()
html_dict = {}
for t in tickers:
    try:
        html_dict[t] = fetch_html(f"https://www.dadosdemercado.com.br/acoes/{t}")
    except:
        continue
```
